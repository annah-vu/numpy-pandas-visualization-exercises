```python
#1.) plot y=x^2 - x + 2 that generates (x,y) points and add (0,0) to the graph

import matplotlib.pyplot as plt
x = list(range(-10,10))
y = [n**2 - n + 2 for n in x]
plt.plot(x,y)
plt.text(0,0, '(0,0)', fontsize=10)
plt.show()
```

```python
#2.) y = sqrt(x)
x = list(range(0,10))
y = [(n**(1/2)) for n in x]
plt.plot(x,y, c="green")
plt.title("y=sqrt(x)")
plt.show()
```

```python
#2.) y = x^3
x = list(range(0,5))
y = [n**3 for n in x]
plt.plot(x,y)
plt.title("y=x^3")
plt.show()
```

```python
#2.) y = 2^x
x = list(range(0,5))
y = [2**n for n in x]
plt.plot(x,y)
plt.title("y=2^x")
plt.show()
```

```python
#2.) y = 1/(x+1)
x = list(range(0,5))
y = [1/(n+1) for n in x]
plt.plot(x,y, c = 'pink')
plt.title("y=1/(x+1)")
plt.show()
```

```python
#3.) Combine the figures you created in the last step into one large figure with 4 subplots.
x = list(range(0,5))
y2 = [(n**(1/2)) for n in x]
y3 = [n**3 for n in x]
y4 = [2**n for n in x]
y5 = [1/(n+1) for n in x]
plt.plot(x, y2, c='orange')
plt.plot(x, y3, c='green')
plt.plot(x, y4, c='blue')
plt.plot(x, y5, c = 'red')
plt.ylim(0,10)
plt.title("All graphs")
plt.show()
```

```python
#4.) Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.
x = list(range(0,5))
y2 = [(n**(1/2)) for n in x]
y3 = [n**3 for n in x]
y4 = [2**n for n in x]
y5 = [1/(n+1) for n in x]
plt.plot(x, y2, c='orange', label = 'y=sqrt(x)')
plt.plot(x, y3, c='green', label = 'y = x^3')
plt.plot(x, y4, c='blue', label = 'y= 2^x')
plt.plot(x, y5, c = 'red', label = 'y = 1/(x+1)')
plt.ylim(0,10)
plt.title("All graphs")
plt.legend()
plt.show()
```
