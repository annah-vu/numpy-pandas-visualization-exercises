```python
#1.) plot y=x^2 - x + 2 that generates (x,y) points and add (0,0) to the graph

import matplotlib.pyplot as plt
x = list(range(-10,11))
y = [n**2 - n + 2 for n in x]
plt.plot(x,y)
plt.text(-.75,0, '(0,0)', fontsize=10)
plt.title('$y=x^2 - x +2$')
plt.xlabel('x')
plt.ylabel('y')
plt.annotate('Origin', xy = (0,0), xytext = (0,100), arrowprops={})
plt.show()
```

```python
#3.) Combine the figures you created in the last step into one large figure with 4 subplots.
plt.figure(figsize = (10,10))
plt.subplot (2,2,1)
x = list(range(0,8))
y1 = [(n**(1/2)) for n in x]
plt.scatter(x,y1, s =30, c='red')
plt.title('$y=\sqrt{x}$')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2,2,2)
x2 = list(range(-2,3))
y2 = [n**3 for n in x2]
plt.scatter(x2, y2, s =30, c='cyan')
plt.title("$y=x^3$")
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2,2,3)
x3 = list(range(-5,6))
y3 = [2**n for n in x3]
plt.scatter(x3,y3, s=30, c='orange')
plt.title("$y=2^x$")
plt.xlabel('x')
plt.ylabel('y')


plt.subplot(2,2,4)
x4 = list(range(-10,-1)) + list(range(0,11))
y4 = [1/(n+1) for n in x4] 
plt.scatter(x4,y4, s=30, c='green')
plt.title("y=1/(x+1)")
plt.xlabel('x')
plt.ylabel('y')


plt.tight_layout()

plt.show()
```

```python
#4.) Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.
plt.figure(figsize = (10,5))
x = list(range(0,5))
y2 = [(n**(1/2)) for n in x]
y3 = [n**3 for n in x]
y4 = [2**n for n in x]
y5 = [1/(n+1) for n in x]
plt.plot(x, y2, c='orange', label = '$y=\sqrt{x}$')
plt.plot(x, y3, c='green', label = '$y = x^3$')
plt.plot(x, y4, c='blue', label = '$y= 2^x$')
plt.plot(x, y5, c = 'red', label = 'y = 1/(x+1)')
plt.ylim(0,10)
plt.xlabel('x')
plt.ylabel('y')
plt.title("All graphs")
plt.legend()
plt.show()
```
