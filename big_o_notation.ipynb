```python
import math as math
```

```python
plt.figure(figsize = (16,8))
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.title('Big O Notation')

x = list(range(1,5))
y1 = [0*n + 1 for n in x]
y2 = [math.log(n) for n in x]
y3 = [n for n in x]
y4 = [n * math.log(n) for n in x] 
y5 = [n**2 for n in x]
y6 = [2**n for n in x]
y7 = [math.factorial(n) for n in x]
y8 = [n**n for n in x]
plt.ylim(0,20)
plt.plot(x,y1, c='red', label = "O(1)")
plt.plot(x,y2, c='orange', label = "$O\log(n)$")
plt.plot(x,y3, c ='cyan', label = "O(n)")
plt.plot(x,y4, c ='green', label = "$O(n\log(n))$")
plt.plot(x,y5, c ="blue", label = "$O(n^2)$")
plt.plot(x,y6, c = "purple", label = "$O(2^n)$")
plt.plot(x,y7, c = 'black', label = "$On!$")
plt.plot(x,y8, c ="magenta", label = "$O(n^n)$")
plt.legend()
plt.show()
```
