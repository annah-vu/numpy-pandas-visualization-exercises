```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data

import seaborn as sns
```

```python
iris_df = data('iris')
iris_df

iris_df = iris_df.rename(columns= {
                                    'Sepal.Length': 'sepal_length',
                                    'Sepal.Width': 'sepal_width',
                                    'Petal.Length':'petal_length',
                                    'Petal.Width': 'petal_width',
                                    'Species': 'species'
})
iris_df
```

```python
#1.) What does the distribution of petal lengths look like?
sns.displot(iris_df.petal_length)
```

```python
#2.) Is there a correlation between petal length and petal width?
sns.relplot(x='petal_length', y='petal_width', data=iris_df)
plt.title('Petal Length vs. Petal Width')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
#There appears to be a correlation
```

```python
#3.) Would it be reasonable to predict species based on sepal width and sepal length?
sns.relplot(x='sepal_length', y='sepal_width', hue = 'species', data=iris_df)
plt.title('Sepal Length vs. Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

#There is not much of a correlation, so no. 
```

```python
#4.) Which features would be best used to predict species?
 #probably petal length and width. It's strongly correlated.
sns.relplot(x='petal_length', y='petal_width', hue = 'species', data=iris_df)
plt.title('Petal Length vs. Petal Width')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
```

```python
#4.) Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each. 
from env import host, username, password
def get_db_url(username,password,host,db_name):
    return f'mysql+pymysql://{username}:{password}@{host}/{db_name}'
get_db_url(username,password,host, 'chipotle')
chipotle_url = get_db_url(username,password,host, 'chipotle')
orders_df = pd.read_sql('select * from orders', chipotle_url)
orders_df.item_price = orders_df.item_price.str.replace('$','').str.replace(',','').astype('float')
```

```python
orders_df.groupby('item_name').quantity.sum().sort_values(ascending = False).nlargest(4, keep='all')
```

```python
ch_top_four = orders_df.groupby('item_name')[['quantity', 'item_price']].sum().sort_values(by = 'quantity', ascending = False).head(4)
ch_top_four
```

```python
ch_top_four.columns = ['Quantity Sold', 'Revenue Made']
ch_top_four
```

```python
sns.barplot(x='Quantity Sold', y='Revenue Made', data = ch_top_four, hue = ch_top_four.index)
```
